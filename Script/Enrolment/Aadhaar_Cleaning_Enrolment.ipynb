```python
def standardize_district(row):
    state = row['standardized_state']
    district = row['district']

    if pd.isna(state) or pd.isna(district):
        return None

    # normalize district text
    d = (
        str(district)
        .lower()
        .replace('&', 'and')
        .replace(r'\s+', ' ')
        .replace('.', '')
        .strip()
    )

    # lookup (state, district)
    key = (state, d)

    if key in district_corrections:
        return district_corrections[key]

    # fallback: cleaned title-case
    return d.title()
```

```python
df['standardized_district'] = df.apply(standardize_district, axis=1)
```
